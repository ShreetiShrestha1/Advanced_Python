## Question 6: List Sanitization and Metrics
Given:
```python
marks = [45, -3, 88, "NA", 102, None, 67]
```

Write a function that:
1. Removes invalid entries  
2. Returns:
   - Cleaned list  
   - Number of invalid entries  
   - Percentage of valid marks  

Return values using a tuple.


````python
def sanitize_marks(marks):
    # this handles empty list to prevent division by zero
    if not marks:
        return [], 0, 0.0

    cleaned = []
    invalid_count = 0

    for m in marks:
        # this checks if item is a number AND within range
        if isinstance(m, (int, float)) and 0 <= m <= 100:
            cleaned.append(m)
        else:
            invalid_count += 1

    total = len(marks)
    # This calculates percentage of data that was usable
    valid_percentage = (len(cleaned) / total) * 100

    return cleaned, invalid_count, round(valid_percentage, 2)

marks = [45, -3, 88, "NA", 102, None, 67]
cleaned_data, discarded, accuracy = sanitize_marks(marks)

print("Cleaned List:",cleaned_data)
print("Invalid Items Removed: ",discarded)
print("Data Integrity: ",accuracy)
````

## Question 8: Prime Number Function
Write `is_prime(n)` and display prime numbers between 1 and 50.

````python
def is_prime(n):
    if n <= 1: #prime numbers are greater than one
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print("Prime numbers between 1 and 50:")
for num in range(1, 51):
    if is_prime(num):
        print(num, end= " ")
````

## Question 9: Function Composition
Write functions to:
1. Calculate total marks  
2. Calculate average  
3. Classify result (Distinction / Pass / Fail)

Then write a function that combines them and returns:
```
(name, total, average, result)
```

Explain why modular design is beneficial.

````python
# Function to calculate the sum of all elements in the list
def total_marks(marks):
    return sum(marks)

# Function to calculate the arithmetic mean
def average_marks(marks):
    if not marks: 
        return 0
    return sum(marks) / len(marks)

# Function to assign a grade category based on the calculated average
def classify(marks):
    avg = average_marks(marks)
    if avg >= 50:
        return "Distinction"
    elif avg >= 30:
        return "First Class"
    elif avg >= 24:
        return "Pass"
    else:
        return "Fail"
    
def main():

    student_name = "Shreeti"
    marks = [89, 88, 76, 67, 100]

    total = total_marks(marks)
    avg = average_marks(marks)
    grade = classify(marks)
    

    print("Student Name:" ,student_name)
    print("Total Marks:", total)
    print("Average Marks:", avg)
    print("Classification:", grade)

main()
````

Modular design breaks programs into smaller, reusable components to enhance clarity and readability. This structure simplifies testing and maintenance by allowing developers to debug or update individual parts without affecting the entire system.

## Question 10: Defensive Input Parsing
Write a function that accepts a raw input string in the format:

```
"Ram|101|22|78, 88, 91"
```
Hint: Parse student data from string format: "Name|Roll|Age|Mark1, Mark2, Mark3"

Tasks:
1. Extract name, roll number, age, and marks  
2. Convert them into appropriate data types  
3. Validate age (≥16) and marks (0–100)  
4. Return the result as a dictionary using roll number as the key  

The function must not crash for malformed input.

````python
def parse_student(input_str):
    try:
        # This splits the main string into its 4 primary components
        parts = input_str.split("|")
        
        # This ensure the string has exactly Name, Roll, Age, and Marks
        if len(parts) != 4:
            return {}

        name = parts[0].strip()
        roll = int(parts[1].strip())
        age = int(parts[2].strip())

        marks = [int(m.strip()) for m in parts[3].split(",")]

        if age < 16 or any(m < 0 or m > 100 for m in marks):
            return {}

        return {roll: {"Name": name, "Age": age, "Marks": marks}}

    except (ValueError, IndexError):
        return {} 

print(parse_student("Shreeti|101|20|66, 88, 99"))#example
````

## Question 11: Type Stability and Conversion
Given the list:
```python
data = [10, "20", 30.5, "abc", None, True]
```

Write a function that:
1. Separates numeric and non-numeric values  
2. Converts numeric values to float  
3. Returns both lists  

Explain in comments why `True` behaves as a numeric value.

````python
data = [10, "20", 30.5, "abc", None, True]

def separate_and_convert(input_list):
# This separates numbers from other data types and converts valid numbers to floats.
    numeric = []
    non_numeric = []

    for item in input_list:
#True behaves like a numeric value because bool inherits from int in Python, so True acts like 1 and False acts like 0.
        if isinstance(item, (int, float)) and not isinstance(item, bool):
            numeric.append(float(item)) 
        else:
            non_numeric.append(item)

    return numeric, non_numeric

numeric_values, non_numeric_values = separate_and_convert(data)

print("Numeric values (floats):", numeric_values)
print("Non-numeric/Filtered values:", non_numeric_values)
````

## Question 13: Set-Based Enrollment Analysis
Given:
```python
python = {"Amit", "Sita", "Hari", "Nita"}
sql = {"Hari", "Ram", "Nita"}
statistics = {"Amit", "Hari", "Gita"}
```

Find:
1. Students enrolled in exactly two subjects  
2. Students enrolled in only one subject  
3. Students enrolled in all subjects

````python
#set of students in differet subjects
python = {"Amit", "Sita", "Hari", "Nita"}
sql = {"Hari", "Ram", "Nita"}
statistics = {"Amit", "Hari", "Gita"}

all_subjects = python & sql & statistics#intersecion

exactly_two = (
    (python & sql) | (python & statistics) | (sql & statistics)) - all_subjects

only_one = (python | sql | statistics) - exactly_two - all_subjects

print("Students enrolled in exactly two subjects:", exactly_two)
print("Students enrolled in only one subject:", only_one)
print("Students enrolled in all subjects:", all_subjects)
````

## Question 14: Dictionary Aggregation Logic
Given:
```python
records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}
```

Write a function that:
1. Computes average per student  
2. Identifies top and bottom performers  
3. Returns results using a single dictionary

````python
records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}

def analyze_performance(records):
    # This calculates average per student using dictionary comprehension
    averages = {name: round(sum(m)/len(m), 2) for name, m in records.items()}

    # Identifys keys with the highest and lowest values
    top = max(averages, key=averages.get)
    bottom = min(averages, key=averages.get)

    # This combine averages and metadata into one dictionary
    return {**averages, "Top Performer": top, "Bottom Performer": bottom}

print(analyze_performance(records))
````

## Question 15: Nested Data Structure Analysis
Given a nested dictionary representing a school's grade book:
```python
school_data = {
    "Class A": {
        "Amit": [78, 82, 91],
        "Sita": [88, 79, 85]
    },
    "Class B": {
        "Hari": [45, 52, 60],
        "Ram": [92, 88, 95]
    }
}
```

Write a function that:
1. Calculates the average marks for each student
2. Finds the top performer in each class
3. Calculates the overall school average
4. Returns a dictionary with:
   - `class_toppers`: {class_name: (student_name, average)}
   - `school_average`: float
   - `student_averages`: {student_name: average}

Explain in comments why nested dictionaries are useful for hierarchical data.

````python
school_data = {
    "Class A": {"Amit": [78, 82, 91], "Sita": [88, 79, 85]},
    "Class B": {"Hari": [45, 52, 60], "Ram": [92, 88, 95]}
}

def analyze_school(data):
    student_averages = {}
    class_toppers = {}
    all_scores = []

    # Nested dictionaries are useful for hierarchical data because they map logical relationships (Class -> Student) into a searchable tree.
    for class_name, students in data.items():
        current_class_averages = {}
        
        for name, marks in students.items():
            avg = sum(marks) / len(marks)
            student_averages[name] = round(avg, 2)
            current_class_averages[name] = avg
            all_scores.extend(marks) 
        
        # this identify topper by finding max value in the class subset
        topper_name = max(current_class_averages, key=current_class_averages.get)
        class_toppers[class_name] = (topper_name, round(current_class_averages[topper_name], 2))

    # this calculate global average
    school_avg = sum(all_scores) / len(all_scores) if all_scores else 0

    return {
        "class_toppers": class_toppers,
        "school_average": round(school_avg, 2),
        "student_averages": student_averages
    }

print(analyze_school(school_data))
````
